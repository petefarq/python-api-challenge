#### Conclusions:
* Temperatures Trend Higher at Lower Latitudes: While there are many factors at play in varying temperatures, at the most basic level the sun's rays hit central regions at closer to a right angle, concentrating energy in a smaller area. Closer to the poles the suns rays hit at an acute angle, spreading energy over a wider area.

* Low Humidity Clustered Around 20 and -20  Latitude: Many of the lowest humidity cities were clustered with 10 degrees of 20 or -20 latitude. This is because our dataset pulled in cities in arid/desert regions that are clustered around those latitudes, including Iran, Namibia, Egypt, Mali, Algeria and central Australia. These areas receive air that has risen from moist equitorial regions, lost its moisture, and descended from the atmosphere as hot, dry air.

* Very High Outlier Temperatures Similarly Clustered Around 20 and -20  Latitude: As with low himidity, the highest temperatures are clustered around the 20 degree latitude bands, particularly in Niger, Iran, Mali, and Algeria. These desert climates are consistently the hottest places on earth, with little cloud cover or surface water to absorb daytime energy from the sun.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import datetime

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1600)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1600)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"

# This dataframe will store responses
weather_df = pd.DataFrame(columns = ['City', 'Cloudiness', 'Country','Date','Humidity','Lat','Lng','Max Temp','Wind Speed'])

citycount = 0

for city in cities:

    # check to see if we have 500 cities with data  yet
    if citycount <=499:
    
        # Build query URL
        query_url = url + "appid=" + api_key + "&q=" + city + "&units=imperial"

        # Get weather data and put into a json
        weather_response = requests.get(query_url)
    
        weather_json = weather_response.json()

        # Test if the city was found (code 200)-- if so, print log and store data if weather_df
        if weather_json['cod'] == 200:
      
            print(f"Data obtained for: {weather_json['name']}, id{citycount}.")
              
            citycount = citycount + 1
   
            # Create a dataframe for the individual city and then append it to weather_df
            city_df= pd.DataFrame(({'City' : weather_json['name'], 
                                    'Cloudiness' : weather_json['clouds']['all'],
                                    'Country' : weather_json['sys']['country'],
                                    'Date': datetime.date.fromtimestamp(int(weather_json['dt'])),
                                    'Humidity': weather_json['main']['humidity'],
                                    'Lat': weather_json['coord']['lat'],
                                    'Lng': weather_json['coord']['lon'],
                                    'Max Temp': weather_json['main']['temp_max'],
                                    'Wind Speed': weather_json['wind']['speed']}), index = [0])

            weather_df = weather_df.append(city_df, ignore_index = True) 
    
         # if city was not found, print the error message
        else:
          
            print("City not found")
              
```

### Export and display the data
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df.to_csv(output_data_file)
weather_df.head(10)
```

#### Latitude vs. Temperature Plot

```python
plt.figure(figsize=(10,8))
plt.scatter(weather_df['Lat'], weather_df['Max Temp'], color='b', alpha=0.75)
plt.title(f"City Latitude Vs. High Temp ({weather_df.iloc[0,3]})")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.savefig("output_data/Lat vs Temp.png")
```
